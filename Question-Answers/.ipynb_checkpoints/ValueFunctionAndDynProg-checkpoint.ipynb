##### Question (a)

Since the utility in period 4 is not a conern, the optimal consumption of the consumer is to spend whatever he has in period 3 and to leave nothing for period 4. So we should expect to have 

$$
c_3 = m_3
$$

To solve it more formally, let's start from the optimizing problem in period 3

$$
\max_{c_3} u(c_3) \\
\text{s.t}\;(m_3 - c_3)R_4 = m_4 \geq 0
$$

First order derivative 

$$
u'(c_3) >0 
$$


Therefore subject to $m_4 \geq 0$, that is $c_3 \leq m_3$, we should have 

$$
c_3 = m_3
$$

And then we can write the value function in period 3.

$$
v_3(m_3) = u(m_3)
$$

##### Question (b)

The consumer chooses $c_2$ to 

$$
\max \{u(c_2) + \beta u(c_3)\} \\
\text{s.t}\ (m_2 -c_2)R_3 = m_3 \\
$$

Besides note that optimization in period 3 already gives us 

$$
c_3 = m_3 
$$

The first order condition for optimization in period 2 requires 

$$
u'(c_2) = \beta R_3 u'(c_3)   \\
\frac{1}{c_2} = \beta \frac{1}{(m_2-c_2)}
$$

Then we solve $c_2$ 

$$
c_2 = \frac{1}{(1+\beta)}m_2
$$

We could also get $m_3$ as the expression of $m_2$

$$
m_3 = (m_2 - c_2)R_3 = \frac{\beta R_3}{1+\beta}m_2
$$

The value function at period 2 is 

$$
v_2(m_2) = \max_{c_2} \{u(c_2) + \beta v_3(m_3)\} \\ 
\text{s.t}\; (m_2 - c_2)R_3 = m_3  
$$

For $v_3(m_3) = u(m_3)$, we have 


$$
v_2(m_2) = \max_{c_2} u(c_2) + \beta u(m_3) \\ 
         = \max_{c_2} \{u(c_2) + \beta u((m_2-c_2)R_3)\}
$$

We could then explicitly write the $v_2(m_2)$ as a function of $m_2$:
$$
v_2(m_2) = u(c_2(m_2)) + \beta u(c_3(m_3)) \\
  = ln(\frac{\beta m_2}{1+\beta}) + \beta ln(m_3) \\
  =  ln(\frac{\beta m_2}{1+\beta}) + \beta ln(\frac{\beta R_3 m_2 }{1+\beta}) \\
  = (\beta+1)ln(m_2) + ln(\frac{\beta}{1+\beta}) +\beta ln(\frac{\beta R_3}{1+\beta})
$$

To verify envolope theorem, 

$$
v'_2(m_2) = \frac{1+\beta}{m_2} \\
\text{and}\; u'(c_2(m_2)) = u'(\frac{1}{1+\beta}m_2) \\
   = \frac{1+\beta}{m_2} \\
$$

So Envolope theorem holds. 

$$
v'_2(m_2) = u'(c_2(m_2))
$$

##### Question (c)

$$
max\; \{u(c_1) + \beta u(c_2) + \beta^2 u(c_3)\} 
$$

The maximized value of  $\beta u(c_2) + \beta^2 u(c_3)$ can be written as a function of $m_2$, that is $v_2(m_2)$ we obtained in question (b) 

Then the question becomes 

$$
\max_{c_1} \; \{u(c_1) + \beta v_2(m_2)\} \\
\text{s.t}\; (m_1-c_1)R_2 = m_2 
$$

That is 

$$
\max_{c_1}  \; \{u(c_1) + \beta v_2((m_1-c_1)R_2)\} 
$$

The first order condition requires 

$$
u'(c_1) = \beta R_2 v_2'(m_2)
$$

Using Envolope theorem, we substitute $v'_2(m_2)$ with $u'(c_2(m_2))$, we have 

$$
u'(c_1) = \beta R_2 u'(c_2(m_2))  \\
\frac{1}{c_1} = \frac{(1+\beta)\beta R_2}{m_2} = \frac{(1+\beta)\beta}{(m_1-c_1)}
$$

Then we solve $c_1$ as

$$
c_1 = \frac{1}{1+\beta+\beta^2}m_1
$$

We then let's verify Envolope Theorem for period 1, that is $v'_1(m_1) = u'(c_1(m_1))$.

We could write explicitly $v_1(m_1)$ as a function of $m_1$. 

$$
v_1(m_1) = u(c_1) + \beta v_2(m_2) \\
         = u(c_1) + \beta v_2((m_1-c_1)R_2)) \\
         = ln(\frac{1}{1+\beta+\beta^2}m_1) + \beta[(\beta+1)ln(\frac{1+\beta}{1+\beta + R_2}m_1 R_2)) + ln(\frac{\beta}{1+\beta}) +\beta ln(\frac{\beta R_3}{1+\beta})] \\
         =(1+\beta +\beta^2) ln(m_1) + \omega(\beta, R_2,R_3) 
$$

where we use $\omega$ to denote the rest part of function not including $m_1$. 

Then we have 

$$
v'_1(m_1) = \frac{1+\beta+\beta^2}{m_1} 
$$

We also have 

$$
u'(c_1(m_1)) = u'(\frac{1}{1+\beta+\beta^2}m_1) =  \frac{1+\beta+\beta^2}{m_1}
$$

Thus we verify that the Envolope Theorem holds for period 1.
