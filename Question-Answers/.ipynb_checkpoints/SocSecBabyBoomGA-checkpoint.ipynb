##### Intuition 

- We start by thinking of the case with no baby boom, namely constant population. Both policies are equivalent.

- In the case in which tax per head is constant,  baby boom generation altogether pays more tax to be enjoyed by a smaller old population. So the benefits to the previous one generation increase. They are better off as they are dividing a bigger pie. In the mean time, the baby generation is worse off on average as the next generation pays the same amount of tax per capita, the total of which is not enough to provide the same level of benefits to each one of baby boom generation.

- In the case of constant benefit policy, in order to sustain the same level of benefits for the previous generation, the baby boom generation only needs to pay less per person. To provide the same level of benefits per capita to the baby boom generation, the next generation needs to pay more per person than they would have to for the old generation of the same population size as themselves. 



##### Some Proof

Consider the benchmark case of constant population (we assumed to be 1) 

$$
L_{t-1} = L_{t} = L_{t+1} \equiv 1
$$

Suppose each generation pays the same amount of tax per head when they are young,

$$
z_{1,t-1}=z_{1,t}=z_{1,t+1} \equiv z_1
$$

Then the total tax paid by the different generation thus the benefits to be enjoyed by the old in the same period is 

$$
Z_{1,t-1} = -Z_{2,t-1} = Z_{1,t} = -Z_{2,t} = Z_{1,t+1} = -Z_{2,t+1} \equiv z_1
$$

Then the benefit enjoyed by the old per head is 

$$
z_{2,t-1}=z_{2,t}=z_{2,t+1} \equiv z_2 = z_1
$$

This proves constant tax per head is equivalent to constant benefit per head. 

With all other conditions unchanged except for the population in period $t$

$$
L_t > 1  
$$

Under constant tax per capita $z_1$, we have 

$$
Z_{1,t} = -Z_{2,t} > Z_{1,t-1} = -Z_{2,t-1} = Z_{1,t+1} = -Z_{2,t+1} \equiv z_1
$$

Focusing on only benefits, divide it by the old population in the same period, we have 

$$
-z_{2,t} > z_1; \\
-z_{2,t+1} < z_1  
$$

This suggests that $t-1$ generation has more benefits than before, while baby boom generation has less benefits than in the benchmark case. Literally speaking, the baby boom generation would have enjoyed more benefits per head if they don't have so many friends in the same generation.


Under constant benefits per capita $z_2$, we have 

$$
-z_{2,t-1} = -z_{2,t} = -z_{2,t+1} \equiv z_2
$$

The total benefits of each generation is thus

$$
-Z_{2,t+1} > -Z_{2,t-1} = -Z_{2,t} \equiv z_2
$$

The total tax for each generation of young people to pay is 

$$
Z_{1,t+1} > Z_{1,t-1} = Z_{1,t} \equiv z_2
$$

Dividing the total benefits by the young population in the same period, we have 

$$
z_{1,t+1} > z_2; \\
z_{1,t} < z_2
$$

This implies that baby boom generation has less tax to pay than in benchmark case, while the generation after them have to pay more tax. 

Therefore, my intuition proves to be right.
