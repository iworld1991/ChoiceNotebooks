### Problem 1: Friedman's Permanent Income Hypothesis

Under the permanent income hypothesis, we assume

$$c_i = p_i$$

Income process is 

$$y_{t,i} = p_{i}+\xi_{i,t}$$

where $\xi_{i,t}$ is a white noise transitory shock with mean zero.

Estimating a Keyesian consumption function

$$
c_i =\alpha_0 + \alpha_1y_i + \epsilon
$$

##### Question a. How would you expect the estimate of $\alpha$ for farmers to be different from workers if we assume variance of transitory shock of farmers is higher?

Additional information is necessary to answer the question. Specifically, we need to know the relative size of transitory shock and permanent income shock for farmers and workers, respectively. 

To show this, denote income Y, permanent component P, $\zeta$ as transitory component, and consumption C.  

The estimate of parameter

$$\widehat{\alpha_1} = \frac{COV (Y,C)}{VAR (Y)}$$

$$= \frac{COV(P+\zeta,C)}{VAR(P+\zeta)}$$

$$= \frac {COV (P, C) + COV(\zeta,C)}{VAR(P+\zeta)}$$

$$= \frac {COV (P, C) + COV(\zeta,C)}{VAR(P)+VAR(\zeta) +2COV(P,\zeta)}$$



According to PIH 
- $COV(P,\zeta) = 0$ 
- $COV(\zeta,C) = 0$ (consumption is unrelated to transitory income.)
- $COV(P,C) = COV(P,P+V) = VAR (P)$

Therefore, $$\widehat{\alpha_1} = \frac{VAR(P)}{VAR(P)+VAR(\zeta)}$$

Denote $VAR(P_{farmers}) = \rho_P VAR(P_{workers})$ and $VAR(\zeta_{farmers}) = \rho_\zeta VAR(\zeta_{workers})$, we need to know $\rho_P$ and $\rho_\zeta$ in order to compare the estimates of $a_1$ for farmers and workers.

#### Simulation 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline  
```

We run three sets of simulations, respectively. Assuming $\sigma^2_{P,{workers}} = 0.01$ and $\sigma^2_{\zeta,{workers}} = 0.02$

  * Scenario 1. When $\sigma^2_{P,{farmers}}=\sigma^2_{P,{workers}} =0.01$ ; $\sigma^2_{\zeta,{farmers}}=3\sigma^2_{\zeta,{workers}} =0.06$ , we expect ${a_{1,{farmers}}} < a_{1,{workers}}$ as $a_{1,{workers}} = \frac{VAR_{P,{workers}}}{VAR_{P,{workers}} + VAR_{\zeta,{workers}}}= \frac{1}{3}$ and $a_{1,{farmers}} = \frac{VAR_ {P,{farmers}}}{VAR_{P,{farmers}}+VAR_{\zeta,{farmers}}}= \frac{1}{7}$
  
  
  * Scenario 2. When $\sigma^2_{P,{farmers}}=10 \sigma^2_{P,{workers}} = 0.1$ ; $\sigma^2_{\zeta,{farmers}}=3\sigma^2_{\zeta,{workers}} = 0.06$, we expect ${a_{1,{farmers}}} >a_{1,{workers}}$ as $a_{1,{workers}} = \frac{VAR_{P,{workers}}}{VAR_{P,{workers}} + VAR_{\zeta,{workers}}}= \frac{1}{3}$ and $a_{1,{farmers}} = \frac{VAR_ {P,{farmers}}}{VAR_{P,{farmers}}+VAR_{\zeta,{farmers}}}= \frac{5}{8}$
  
  
  * Scenario 3. $\sigma^2_{P,{farmers}}=0.1 \sigma^2_{P,{workers}} = 0.001$ ; $\sigma^2_{\zeta,{farmers}}=3\sigma^2_{\zeta,{workers}} =0.06$, we expect ${a_{1,{farmers}}} < a_{1,{workers}}$ as  $a_{1,{workers}} = \frac{VAR_{P,{workers}}}{VAR_{P,{workers}} + VAR_{\zeta,{workers}}}= \frac{1}{3}$ and $a_{1,{farmers}} = \frac{VAR_ {P,{farmers}}}{VAR_{P,{farmers}}+VAR_{\zeta,{farmers}}}= \frac{1}{61}$ 

```python
### We start by defining a consumption function under the assumption of PIH

np.random.seed(112)

def consumption_func(P): # P is permanent income 
    n = len(P)          # number of periods 
    V = np.random.normal(0,.001,n)  # i.i.d. shock to consumption
    return P+V
```

```python
#### Set common parameters 

nb_obs = 50
a = np.linspace(-10,10)

### For workers
sigma_P_w = np.sqrt(.01)
sigma_Z_w = np.sqrt(.02) 

mean_P_w = 1    # Workers: Average Permanent
mu_P_w = mean_P_w # Workers: Parameters of Distribution
mu_Z_w = 0      # Workers: Average Transitory

mean_P_f = 1     # Farmers: Average Permanent
mu_P_f = mean_P_f     # Farmers: Parameters of Distribution
mu_Z_f = 0       # Farmers: Average Transitory 

### Loop over three of these parameters

sigma_P_ratio_lst = [1,10,.1]
sigma_Z_ratio_lst = [3,3,3]
```

```python
# Loop Over Scenario 1, 2 and 3

fig, axs = plt.subplots(1, 3, sharex=True, figsize=(15,5))
fig.text(0.5, 0.04, 'Y: Income', ha='center')
fig.text(0.08, 0.5, 'C: Consumption', va='center', rotation='vertical')

for id in range(len(sigma_P_ratio_lst)):
    sigma_P_ratio = sigma_P_ratio_lst[id]
    sigma_Z_ratio = sigma_Z_ratio_lst[id]
    
    ### For farmers
    sigma_P_f = sigma_P_w *np.sqrt(sigma_P_ratio)
    sigma_Z_f = sigma_Z_w *np.sqrt(sigma_Z_ratio)
        
    #### Generate data points 
    
    ### Workers
        
    P_w = np.random.normal(mu_P_w,sigma_P_w,nb_obs)
    Z_w = np.random.normal(mu_Z_w,sigma_Z_w,nb_obs)
    
    Y_w = P_w + Z_w
    
    C_w = consumption_func(P_w)
    
    ### Farmers 
        
    P_f = np.random.normal(mu_P_f,sigma_P_f,nb_obs)
    Z_f = np.random.normal(mu_Z_f,sigma_Z_f,nb_obs)
    
    Y_f = P_f + Z_f 
    C_f = consumption_func(P_f)
    
    ### Estimate 
    
    slope_w, intercept_w, r_value_w, p_value_w, std_err_w = stats.linregress(Y_w,C_w)
    line_w = slope_w*a+intercept_w
    slope_f, intercept_f, r_value_f, p_value_f, std_err_f = stats.linregress(Y_f,C_f)
    line_f = slope_f*a+intercept_f
    
    ### Plot
    axs[id].axis([0,2,0,2])
    axs[id].plot(Y_w,C_w,'ro',label='Workers: Sample')
    axs[id].plot(a,line_w,'r-',label='Workers: Estimates')
    axs[id].plot(Y_f,C_f,'bv',label='Farmers: Sample')
    axs[id].plot(a,line_f,'b-',label='Farmers: Estimates')
    axs[id].plot(a,a,'g--',label='PIH Prediction')
    if id==2:
        axs[id].legend(loc='best',fontsize=10)
    axs[id].set_title('Scenario'+str(id+1)+ '\n' '$\sigma^2_{P,{farmers}}=$' +str(sigma_P_ratio_lst[id]) \
                      + '$\sigma^2_{P,{workers}}$' + '\n' + '$\sigma^2_{\zeta,{farmers}}=$' + \
                      str(sigma_Z_ratio_lst[id]) + '$\sigma^2_{\zeta,{workers}}$')
    axs[id].text(1.2, .2, '$\widehat a_{1,{workers}}=$'+str(round(slope_w,2)) \
                 +'\n' +'$\widehat a_{1,{farmers}}=$'+str(round(slope_f,2)) , fontsize=12, color = 'r')
```

### Conclusion 

The simulated results by and large confirm our prediction. But it is important to note that the simulation results are more robust for bigger number of observations. 

##### Question b Saving Rate and Income Growth 

According to PIH, $C_t = P_t$ regardless of realized change of transitory income. Then the saving rate is only related to change of transitory income. In the absense of uncertainty, positive(negative) income change in the next period gives consumers less(more) incentive to save in the current period.

###### Math

a. $Y_t = P_t + \zeta_t $  

b. $C_t = P_t$ 

c. $Y_{t+1} = P_{t+1} + \zeta_{t+1}$

d. $C_{t+1} = P_{t+1}$ 

e. $P_{t+1} = P_{t}$ ( by assumption)

f. $s_t = 1-\frac{C_t}{Y_t}$ (saving rate at time t)

g. $\widehat {Y_{t+1}}=\frac{Y_{t+1} - Y_t}{Y_t}$

a, c, e and g $\Rightarrow \widehat Y_{t+1}=\frac{\zeta_{t+1}-\zeta_t}{P_t + \zeta_t}$

a, b and f $\Rightarrow s_t = 1- \frac{P_t}{P_t + \zeta_t} = \frac{\zeta_t }{P_t + \zeta_t}$

$\Rightarrow s_t = -\widehat Y_{t+1} + \frac{\zeta_{t+1}}{{P_t + \zeta_t}}$

This implies that saving rate is negatively correlated with income change from period $t$ to $t+1$, with a coefficient being -1.

#### Simulation

- Following problem 1, we assume the variance of permanent and transitory income to be 0.01 and 0.02, respectively.

```python
#### Set common parameters 

nb_obs = 100
a = np.linspace(-1,1)

sigma_P = np.sqrt(.01)
sigma_Z = np.sqrt(.02) 
mu_P = 1    # Workers: Average Permanent
mu_Z = 0      # Workers: Average Transitory

#### Period t 

P_t1 = np.random.normal(mu_P,sigma_P,nb_obs)
Z_t1 = np.random.normal(mu_Z,sigma_Z,nb_obs)

Y_t1 = P_t1 + Z_t1
C_t1 = consumption_func(P_t1)
s_t1 = 1-C_t1/Y_t1

#### Period t+1

P_t2 = P_t1
Z_t2 = np.random.normal(mu_Z,sigma_Z,nb_obs)

Y_t2 = P_t2 + Z_t2
C_t2 = consumption_func(P_t2)

### t to t+1

Y_chg = (Y_t2 - Y_t1)/Y_t1
```

```python
plt.axis=([-1,-1],[1,1])
plt.plot(s_t1, Y_chg,'ro',label='Observations')
plt.plot(s_t1,-s_t1,label='PIH Prediction')
plt.axhline(y=0,linestyle='-',color='k')
plt.axvline(x=0,linestyle='-',color='k')
plt.xlabel('Saving Rate')
plt.ylabel('Income Change')
plt.legend(loc=1)
plt.title('Income Change and Saving Rate')
```

#### Conclusion

The simulation results confirm our prior regarding negative correlation of income change and saving rate. This is intuitive. 

##### Question c. Effects of Tax Withholding on Consumption

- Based on PIH, consumption responds to permanent income but not to transitory income. One need to consider if the implemented policy changed permanent income of households to decide if it would stimulate consumption.   

- As the reduction in tax withholding in current period would be offset by a same amount of increase in later period, it did not change permanent income profile of the households. Thus in theory it has no impact on consumption. 

- The prediction hinges on the assumption of perfect rationality of consumers. Specifically, consumers did not have difficulty recognizing their permanent income would not change in face of the policy annoucement. Also, it implicitly assume that consumers are not liquidity constrained. 

- There are some empirical literature testing the effect of the policy. For instance, Shapiro and Slemrod (1993) found that actually a proportion of households changed consumption in response to the policy, implying a deviation from standard PIH theory.
