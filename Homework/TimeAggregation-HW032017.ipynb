##### Problem 1 The Time Aggregation Problem 

Consumption is surveyed annually and denoted $C_t$ for year $t$.

But consumers make decisions semiannually. For year t, $C_\tau$ is the first half year and $C_{\tau+1}$ is the consumption in second half year. It follows a random walk:   

$$
C_{\tau+1} = C_{\tau} + \epsilon_{\tau+1}; \quad \mathbb{E}_t[\epsilon_{t+n}] = 0 \;\forall n > 0 \quad (1)
$$




##### Question (a) 

The measured consumption (annual) is denoted as below 

$$
\Delta \bar C_t = \underbrace{C_{\tau} + C_{\tau+1}}_{\bar C_t} - \underbrace{(C_{\tau-1}+ C_{\tau-2})}_{\bar C_{t-1}} \quad (2)
$$ 

Substituting $C_{\tau-2} = C_{\tau-1}- \Delta C_{\tau-1}$, $C_{\tau+1} = C_{\tau}+\Delta C_{\tau+1}$ into the equation, we have 

$$
\Delta \bar C_t = C_{\tau} + C_{\tau}+\Delta C_{\tau+1} - C_{\tau-1}- C_{\tau-1}+ \Delta C_{\tau-1} \\
= 2(C_{\tau}-C_{\tau-1}) + \Delta C_{\tau+1} + \Delta C_{\tau-1} \\
= \Delta C_{\tau+1} + 2 \Delta C_{\tau} + \Delta C_{\tau -1}  \quad (3)
$$

##### Question (b)

Utilizing the random walk assumption, we could further write $\Delta \bar C_t$ as below

$$
\Delta \bar C_t = \epsilon_{\tau+1} + 2 \epsilon_{\tau} + \epsilon_{\tau-1}  \quad (4)
$$

Similarly, we could write $\Delta \bar C_{t-1}$

$$
\Delta \bar C_{t-1} = \epsilon_{\tau-1} + 2 \epsilon_{\tau-2} + \epsilon_{\tau-3} \quad (5)
$$

From (4) and (5) we have 

$$
Cov(\Delta \bar C_t,\Delta \bar C_{t-1}) = E[(\epsilon_{\tau+1} + 2 \epsilon_{\tau} + \epsilon_{\tau-1})(\epsilon_{\tau-1} + 2 \epsilon_{\tau-2} + \epsilon_{\tau-3})] \\
= E(\epsilon_{t-1}^2) \\
= Var(\epsilon_{t-1}) \neq 0
$$

Therefore, $\Delta \bar C_t$ and $\Delta \bar C_{t-1}$ are correlated with each other. This would not be the case if the consumers make decision on an annual basis, in which the annual change from $t-1$ to $t$ and from $t-2$ to $t-1$ is uncorrelated.

#### Some Derivations 

Under certainty equivalence preference and other assumptions made by Hall(1978), the consumers maximize its expected life-long utility:

$$
\mathbb{E}_\tau\left[\sum^\infty_{n=0} R^{-n}c_{\tau+n}\right] = b_\tau + \mathbb{E}_\tau \left[\sum^\infty_{n=0} R^{-n} y_{\tau+n}\right]  \quad (8)
$$

Please be noted that we use $\tau$ as subscript of variables as the consumers make decision semi-annually.

Assuming an income process composed of both permanent and transitory income:

$$
P_{\tau+1} = P_{\tau} + \psi_{\tau+1}, \quad \mathbb{E}(\psi_{\tau+n}) =0 \; \forall n >0 \quad \\
Y_{\tau+1} = P_{\tau+1} + \theta_{\tau+1} \quad \mathbb{E}(\theta_{\tau+n}) =0 \; \forall n >0 \quad
$$

Assuming $b_\tau =0$, then we could obtain from (8) that 

$$
c_\tau =\frac{r}{R}\theta_\tau + p_\tau  \quad (9)
$$

This suggests that consumption is a function of both transitory and and permanent income shocks. With reasonable assumption that $r \rightarrow 0$, the marginal propensity to consume with respect to transitory income shock is close to zero, while the marginal propensity to consume out of permanent income is 1. We could simplify the consumption function as 

$$
c_\tau = p_\tau   \quad (10) 
$$

We use this in our simulation to generate our semi-annual consumption series.

##### Simulation

```python
### Import python modules 

import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
%matplotlib inline  
```

```python
## Define a consumption function

def cons_func(p_t):
    return p_t
```

```python
## Define realized income and consumption 
## Parameters 
#### p0: initial permanent income
#### periods: number of periods
#### sigma_psi standard deviation of permanent income shock
#### sigma_theta standard deviation of transitory income shock

def inc_con_process(p0,periods,sigma_psi,sigma_theta):   
    y = np.empty(periods)
    p = np.empty(periods)
    c = np.empty(periods)
    p[0] = p0
    # c[0] = cons_func(p[0])
    psi = np.empty(periods)
    theta = np.empty(periods)
    for i in range(periods-1):
        p[i+1] = p[i]+ np.random.normal(0,sigma_psi,1) 
        y[i+1] = p[i+1] + np.random.normal(0,sigma_theta,1)
        c[i+1] = cons_func(p[i+1])
    return y[1:-1],c[1:-1]       # Give both income and consumption
```

```python
### Define function that gives measured income and consumption on annual basis 

def aggregated(series):
    nb_semi = len(series)  ## Number of semi periods
    nb_ann  = int(nb_semi/2)   ## Number of years 
    series_bar = np.empty(nb_ann)   ## Series of measured income
    for i in range(nb_ann):    
        series_bar[i] = series[2*i] + series[2*i+1]
    return series_bar
```

#### How does the consumption and income evolve?

```python
fig, ax = plt.subplots(1, 2, sharex=True, figsize=(15,5))

###  Test 1 
y_test,c_test= inc_con_process(100,1000,.1,.2)
ax[0].axis([0,1000,95,105])
ax[0].plot(y_test,label='$y$',linestyle='-',color='b')
ax[0].plot(c_test,label='$c$',linestyle='--',color='r')
ax[0].set_title('Sample Series of Income and \n Consumption  (Semi annual)')
ax[0].legend(loc=2)

### Test 2 
y_bar_test = aggregated(y_test)
c_bar_test = aggregated(c_test)
ax[1].axis([0,500,180,220])
ax[1].plot(y_bar_test,label='$y$',linestyle='-',color='b')
ax[1].plot(c_bar_test,label='$c$',linestyle='--',color='r')
ax[1].set_title('Sample Series of Measured Income \n and Consumption (Annual)')
```

```python
#### How does the Change in consumption and income evolve?
```

```python
# Calculate changes and run the regression

delta_y_test = y_test[:-1] - y_test[1:]
delta_c_test = c_test[:-1] - c_test[1:]

delta_y_bar_test = y_bar_test[:-1] - y_bar_test[1:]
delta_c_bar_test = c_bar_test[:-1] - c_bar_test[1:]

alpha1_test = stats.linregress(delta_c_test[:-1],delta_y_test[1:])[0]
alpha1_bar_test = stats.linregress(delta_c_bar_test[:-1],delta_y_bar_test[1:])[0]

fig, ax = plt.subplots(1, 2, sharex=True, figsize=(15,5))

### Test 3 Changes using Measured Semi-annual data

ax[0].axis([0,500,-1.5,1.5])
ax[0].plot(delta_y_test[1:])
ax[0].plot(delta_c_test)
ax[0].set_title('Changes of Income \n and Consumption (Semi-annual)')
ax[0].text(300,-1,'$\widehat a_1=$'+str(round(alpha1_test,2)),size='large')

### Test 4 Changes using Measured Annual data

ax[1].axis([0,500,-1.5,1.5])
ax[1].plot(delta_y_bar_test[1:])
ax[1].plot(delta_c_bar_test)
ax[1].set_title('Changes of Income \n and Consumption (Annual)')
ax[1].text(300,-1,'$\widehat a_1=$'+str(round(alpha1_bar_test,2)),size='large')
```

We could see from the charts above that change of income and consumption are not correlated on semi-annual frequency, while positively correlated on annual basis. 

Now, we generate income and consumption series based on four combinition of $Var(\theta_\tau)$ and $Var(\psi_\tau)$ 

1. $Var(\psi_\tau) = 0.1^2 \quad and \quad Var(\theta_\tau) = 0.1^2$
2. $Var(\psi_\tau) = 0.1^2 \quad and \quad Var(\theta_\tau) = 0.2^2$
3. $Var(\psi_\tau) = 0.2^2 \quad and \quad Var(\theta_\tau) = 0.1^2$
4. $Var(\psi_\tau) = 0.2^2 \quad and \quad Var(\theta_\tau) = 0.2^2$

And we report estimates of $\alpha_1$ in below.

```python
## Simulated Regression based on Different Income Process 

#### Set invariable parameters 

nb_obs = 1000
p0 = 100 

#### Set variable parameters 

sigma_psi_list = [.1,.2]     # Std of permanent income shocks
sigma_theta_list = [.02,.05,.1,.2]  # Std of transitory income shocks

for sigma_psi in sigma_psi_list:
    for sigma_theta in sigma_theta_list:
        y,c = inc_con_process(p0,nb_obs,sigma_psi,sigma_theta)
        y_bar = aggregated(y)
        c_bar = aggregated(c)
        delta_y_bar = y_bar[:-1] - y_bar[1:]
        delta_c_bar = c_bar[:-1] - c_bar[1:]
        slope_w = stats.linregress(delta_c_bar[:-1],delta_y_bar[1:])[0] 
        print('Estimate of alpha1 is '+str(round(slope_w,2)) + ' , when sigma_psi = '\
              + str(sigma_psi)+ ', sigma_theta ='+ str(sigma_theta))
        
```

##### Some extension 

In order to see why more clearly, let's assume the income in each period is composed of both a permanent and transitory component, in which the permanent income component follows the random walk. 

$$
P_{\tau+1} = P_{\tau} + \psi_{\tau+1}, \quad \mathbb{E}(\psi_{t+n}) =0 \; \forall n >0 \\
Y_{\tau+1} = P_{\tau+1} + \theta_{\tau+1} \quad \mathbb{E}(\theta_{t+n}) =0 \; \forall n >0 
$$


Also we assume marginal propensity to consumption out of permanent income is equal to 1, and that out of transitory income is equal to 0. That is 

$$ 
C_\tau = P_\tau \\
C_\tau = P_{\tau-1} + \psi_\tau \\
\Delta C_\tau = \psi_\tau - \psi_{\tau-1}
$$

Therefore, we have 

$$
\Delta Y_{\tau} = P_{\tau} + \psi_{\tau} - P_{\tau-1} - \psi_{\tau-1} \\
 = P_{\tau-1} + \psi_{\tau} + \psi_{\tau} - P_{\tau-1} - \psi_{\tau-1} \\
 = \psi_{\tau} + \theta_{\tau} -\theta_{\tau-1} 
$$


We could write $\Delta \bar Y_{t-1}$ as 


$$
\Delta \bar Y_{t-1} = \psi_{\tau-1} + \theta_{\tau-1}-\theta_{\tau-2}   + 2 (\psi_{\tau-2} + \theta_{\tau-2}-\theta_{\tau-3}) + \psi_{\tau-3} + \theta_{\tau-3}-\theta_{\tau-4} \\
=  \psi_{\tau-1} +  2 \psi_{\tau-2} +\psi_{\tau-3} +\theta_{\tau-1} + \theta_{\tau-2}+ \theta_{\tau-3}-\theta_{\tau-4}
$$

We rewrite (3), that is 

$$
\Delta \bar C_t = \Delta C_{\tau+1} + 2 \Delta C_{\tau} + \Delta C_{\tau -1} \\
                = \psi_{\tau+1}-\psi_{\tau} + 2( \psi_\tau - \psi_{\tau-1}) + \psi_{\tau-1} - \psi_{\tau-2} \\
                = \psi_{\tau+1} + \psi_{\tau} - \psi_{\tau-1} - \psi_{\tau-2} 
$$

Then, finally, we could clearly see 

$$
Cov(\Delta \bar Y_{t-1}, \Delta \bar C_t) = E[(\psi_{\tau-1} +  2 \psi_{\tau-2} +\psi_{\tau-3} +\theta_{\tau-1} + \theta_{\tau-2}+ \theta_{\tau-3}-\theta_{\tau-4})(\psi_{\tau+1} + \psi_{\tau} - \psi_{\tau-1} - \psi_{\tau-2}] \\
= E(\psi_{\tau-1}^2) = Var (\psi_{\tau-1}) \neq 0  
$$



#### Some further Comments on Random Walk Theory  

- The random walk theory by Hall(1988) does not rely on specific assumption about income process and relationship between income and consumption. It only provides a simple prediction that the information known in the current period (including income and all other factors) have no predictive power for the consumption change in next period. It, however, does not imply consumption is not to respond to realized income shocks. 

- Under constant certainty equivalence and the assumption $\beta R= 1$, we could solve for the consumption function with respect to income. (See Equation 13 in the lecture note "Consumption Function and PIH".) Consumption should be responsive to both permanent and transitory income shocks. But the estimates of the response varies with the relative size of permanent and transitory income shocks.
